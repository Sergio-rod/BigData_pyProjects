```python

from pymongo import MongoClient as mc
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import (datasets,model_selection as skms, neighbors, metrics as mt, linear_model as lm)
import sqlite3, pandas as pd
import pandas as pd
```

```python

# Ruta al archivo JSON


# Cargar el archivo JSON en un DataFrame de Pandas
df = pd.read_json('tienda.json')
```

```python
print(df)
```

```python
x= df.iloc[:,:-1]
print(x)
```

```python
y= df.iloc[:,-1:]
print(y)
```

```python
corr=df.corr()

print(corr)
```

```python
x_train,x_test,y_train,y_test = skms.train_test_split(x,y,test_size= .5, random_state=42)
```

```python
print(x_train)
print(y_train)
print(x_test)
print(y_test)
```

```python
"""Prueba modelo de regresión"""
#clasificador knn
model = lm.LinearRegression().fit(x_train,y_train)
#entrenamiento


#mandamos datos a predecir e imprimimos 

predict = model.predict(x_test)
print("Valores predichos: \n \n ",predict)

print("Valores reales \n \n",y_test)


```
