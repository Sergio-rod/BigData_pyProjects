```python
from sklearn import (datasets,model_selection as skms, neighbors, metrics)
import pandas as pd
iris= datasets.load_iris()
```

```python
iris
```

```python
print(iris.DESCR)
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_df
```

```python
iris_df['tgt']=iris.target
```

```python
iris_df
```

```python
x=iris_df.iloc[:,:-1]
print(x)
```

```python
y=iris_df.iloc[:,-1].values
print(y)
type(y)
```

```python
x_train,x_test,y_train,y_test = skms.train_test_split(x,y,test_size= .25,stratify=y,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
"""Pruebas"""
#clasificador knn
model = neighbors.KNeighborsClassifier()
#entrenamiento
model.fit(x_train,y_train)

#mandamos datos a predecir e imprimimos 

predicciones= model.predict(x_test)


j=metrics.accuracy_score(predicciones,y_test)
i=metrics.confusion_matrix(predicciones,y_test)



print(j)
print(i)





```
