```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread('prueba.png')
img = imagen.copy()
img.shape
```

```python
### resize
img_size = img.shape
im_size_min = np.min(img_size[0:2])
im_size_max = np.max(img_size[0:2])

im_scale = float(600) / float(im_size_min)
print(im_scale)
```

```python
if np.round(im_scale * im_size_max) > 1200:
    im_scale = float(1200) / float(im_size_max)
new_h = int(img_size[0] * im_scale)
new_w = int(img_size[1] * im_scale)

new_h = new_h if new_h // 16 == 0 else (new_h // 16 + 1) * 16
new_w = new_w if new_w // 16 == 0 else (new_w // 16 + 1) * 16

re_im = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
v1,v2 = (new_h / img_size[0], new_w / img_size[1])
print(v1)
print(v2)
print(re_im.shape)
```

```python
# Points
im = re_im.copy()
im = cv2.rectangle(im,(100,100),(250,180),(255,0,0),4)
#im = cv2.circle(im,(432,113),5,(255,0,0),4,-1)
#im = cv2.circle(im,(704,113),5,(255,0,0),4,-1)
#im = cv2.circle(im,(704,138),5,(255,0,0),4,-1)
#im = cv2.circle(im,(432,138),5,(255,0,0),4,-1)
plt.imshow(im)
```
