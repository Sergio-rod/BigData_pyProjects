```python
X_train, X_test, y_train, y_test = X[:1000], X[1000:2000], y[:1000], y[1000:2000]
```

```python
"""Pruebas"""
#clasificador knn
model = neighbors.KNeighborsClassifier()
#entrenamiento
model.fit(X_train,y_train)

#mandamos datos a predecir e imprimimos 

predicciones= model.predict(X_test)


j=mt.accuracy_score(predicciones,y_test)
i=mt.confusion_matrix(predicciones,y_test)



print(j)
print(i)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1, as_frame=False)
X, y = mnist.data, mnist.target

# Crear un DataFrame a partir de los datos
df = pd.DataFrame(data=X, columns=mnist.feature_names)

# Convertir y a enteros
y = y.astype(int)

# Crear un pairplot
sns.pairplot(df.sample(1000), vars=df.columns[:4])
plt.show()

# Crear un scatterplot
plt.scatter(df['pixel1'], df['pixel2'], c=y, cmap='viridis')
plt.xlabel('pixel1')
plt.ylabel('pixel2')
plt.colorbar()
plt.show()
```
